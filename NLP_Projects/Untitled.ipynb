```python
from tensorflow.keras.preprocessing.text import one_hot
```

- converting in one hot representation

```python
sent=['the glass of milk',
     'the glass of juice',
     'the cup of tea',
     'I am good boy',
     'I am good developer',
     'understand the meaning of words',
     'your videos are good']
sent
```

```python
voc_size=10000   # Vocablary size
```

```python
onehot_repr=[one_hot(word,voc_size) for word in sent] 
# This will give index of each word in the sentence. But index will come in term of 10000. 
print(onehot_repr)
```

```python
## Pass one hot representation to Embeding layer of Keras to convert into Embedding Matrix
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_len=8
#Pad sequence is used to make the all the sentence of same length
embeded_docs =pad_sequences(onehot_repr,padding='pre',maxlen=sent_len)
```

```python
dim=10
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embeded_docs))
```
